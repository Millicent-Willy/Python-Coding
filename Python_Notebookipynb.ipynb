**Z SCORE**

Is a measurement taht describes the distance a value is from the mean of a normal data distribution.

Z score is measured in terms of a standard deviation.

It typically tells us how far a value is in standard deviations so that we can understand it in a scaled value.

Z Score=(X-population mean)/population std deviation


```python
mean=df['male_smokers'].mean()
std=df['male_smokers'].std()
```

```python
df['Z score']= (df['male_smokers']-mean)/std
df['Z score']
```
